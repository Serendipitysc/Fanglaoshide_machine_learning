```python
from sklearn.datasets import fetch_openml
import random

mnist = fetch_openml('mnist_784', version=1)
X = mnist["data"]
Y = X[:1000]
print(type(Y))
```

```python
import numpy as np
#Randomly initialize the index of the starting centroids
def init_centroid(k,data):
    centriods = []
    for x in range(k):
        centriods.append(data[random.randint(0,len(data)-1)])
    return np.array(centriods)


#Function returns the euclidean distance between two points
def euclidean_distance(x,y):
    diff_sq=[0]*784
    for i in range(784):
        diff_sq[i]=(x[i]-y[i])**2
    return sum(diff_sq)**(0.5)

#Return the centroids in clusters
def find_centroids(cluster):
    center=[]
    if len(cluster)==0:
        return center
    x=[]
    for i in range(784):
        x=[]
        for instance in cluster:
            assert (len(instance) != 0)
            x.append(instance[i])
        center.append(sum(x)/len(x))
    return center

#Return the loss function result
def loss_result(clusters, centroids):
    assert(len(clusters) == len(centroids))
    dist = []
    for i in range(len(centroids)):
        for entry in clusters[i]:
            dist.append(euclidean_distance(entry, centroids[i])**2)
    return sum(dist)

#function that compares the difference between two list of lists and return bool values 
def is_different(x, y):
    assert (len(x) == len(y))
    for i in range(len(x)):
        if not np.array_equal(x[i], y[i]):
            return True
    return False
                
#Return 
def my_kmeans(data_matrix, k, n):
    #list to keep track of the loss function values
    loss_function_values = []
    final_loss_function_values = []
    final_centroids = []
    final_clusters_assignment = []
    for i in range(n):
        print ("i=",i)
        local_loss_function_values = []
        #initialize the starting centroid 
        new_centroids = init_centroid(k, data_matrix)
        old_centroids = [0 for i in range(k)]
        old_centroids = np.array(old_centroids)
        #Initialize the old clusters
        old_clusters = [[] for i in range(k)]
        old_clusters = np.array([np.array(xi) for xi in old_clusters])
        
        #assign the membership of the point base on the initialized centroids
        #initialize the first new clusters with empty list of list
        new_clusters = [[] for i in range(k)]
        for entry in data_matrix:
            #list store euclidean distance between one pt in matrix to all the centroid we intialized
            e_dist = []
            for c in new_centroids:
                e_dist.append(euclidean_distance(entry, c))
            #append the data entry to corresponding cluster base on the index of mini of euclidean distance
            new_clusters[e_dist.index(min(e_dist))].append(entry)
        new_clusters = np.array([np.array(xi) for xi in new_clusters])
        print(type(new_clusters[0][0]))
        local_loss_function_values.append(loss_result(new_clusters,new_centroids))
        
        #check for the convergence
        print(is_different(old_clusters, new_clusters))
        print(not np.array_equal(new_centroids, old_centroids))
        
        while (is_different(old_clusters, new_clusters)) and (not np.array_equal(new_centroids, old_centroids)):
            print("inside the while loop")
            #keep a copy of old clusters and centroids
            old_clusters = np.copy(new_clusters)
            old_centroids = np.copy(new_centroids)
            #Find the new centrods
            new_centroids = []
            for cluster in new_clusters:
                new_centroids.append(find_centroids(cluster))
            new_centroids = np.array(new_centroids)
            
            #reassign the membership of the entry in data_matrix
            new_clusters = [[] for i in range(k)]
            for entry in data_matrix:
                #list store euclidean distance between one pt in matrix to all the centroid we intialized
                e_dist = []
                for c in new_centroids:
                    if len(c) == 0:
                        e_dist.append(float('inf'))
                    e_dist.append(euclidean_distance(entry, c))
                #append the data entry to corresponding cluster base on the index of mini of euclidean distance
                new_clusters[e_dist.index(min(e_dist))].append(entry)
            new_clusters = np.array(new_clusters)
            #keep track of the local loss function values
            local_loss_function_values.append(loss_result(new_clusters,new_centroids))
            print(is_different(old_clusters, new_clusters))
            print(not np.array_equal(new_centroids, old_centroids))
            
    
        #keep track of the final loss function values
        loss_function_values.append(local_loss_function_values)
        final_loss_function_values.append(min(local_loss_function_values))
        final_centroids.append(new_centroids.copy())
        final_clusters_assignment.append(new_clusters.copy())
            
            
        
    return min(final_loss_function_values), loss_function_values[final_loss_function_values.index(min(final_loss_function_values))], final_loss_function_values     
            

my_kmeans(Y,10,15)
```
