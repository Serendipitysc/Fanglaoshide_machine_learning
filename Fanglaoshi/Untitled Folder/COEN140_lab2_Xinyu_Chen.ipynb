```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
'''
Function to test the correct format of list of list

Return: Boolen
'''
def test_dataset(data):
    if len(data) != 150:
        return False
   
    for row in data:
        if len(row) != 5:
            return False
       
        for column in row[:-1]:
            if type(column) != np.float64:
                print("not float is a ", type(column), column)
                return False
           
        if type(row[-1]) != str:
            print("not a string")
            return False
   
    return True

data = []

'''
problem1 function read the data, and format the data to numpy array

Return: the numpy array
'''
def problem1():

    ''''load data from .data file to list of list'''
    f=open("iris.data")
    sample=f.readlines()
    for i in range(len(sample)):
        sample[i] = sample[i].strip().split(',')

    for item in sample:
        list=[]
        for i in range(len(item)):
            if i!=(len(item)-1):
                list.append(np.float64(item[i]))
            else:
                list.append(item[i])
        data.append(list)

    print("the test_dataset function returns:",test_dataset(data))
    '''convert data to numpy array'''  
    prepared_data = np.array(data)
    
    return prepared_data

'''
Function parse the data into numpy arrays of data and labels:

Return the numpy data and labels
'''   
def parse_data(data):
    training_data = []
    testing_data = []
    training_data_labels=[]
    testing_data_labels=[]
    '''separate training data and testing data into list of list forms'''
    for i in range(len(data)):
        if i in range(0,40) or i in range(50,90) or i in range(100, 140):
            training_data.append(data[i])
        else:
            testing_data.append(data[i])


    #print(training_data,"\n\n\n\n" ,testing_data)

    '''separate the labels and data in list of list forms'''
    for ii in range(len(training_data)):
        training_data_labels.append(training_data[ii][-1])
        training_data[ii]=training_data[ii][:len(training_data[ii])-1]


    for jj in range(len(testing_data)):
        testing_data_labels.append(testing_data[jj][-1])
        testing_data[jj]=testing_data[jj][:len(testing_data[jj])-1]


    '''convert everything into numpy array'''
    training_data = np.array(training_data)
    training_data_labels = np.array(training_data_labels)
    testing_data = np.array(testing_data)
    testing_data_labels = np.array(testing_data_labels)
    print("No column has been removed:")
    return training_data,training_data_labels,testing_data,testing_data_labels
```

```python
'''
Function problem2 train the data with LDA

Print out testing and training error 

Return the LDA object
'''
def problem2():
    '''LDA test with testing data'''
    lda = LDA()
    lda_test_re = lda.fit(training_data,training_data_labels).score(testing_data,testing_data_labels)
    lda_train_re = lda.fit(training_data,training_data_labels).score(training_data,training_data_labels)
    print("LDA testing error rate is", (1-lda_test_re)*100,"%")
    print("LDA training error rate is", (1-lda_train_re)*100,"%")
    return lda
```

```python
 '''
 Function problem3 train the data with QDA
 
 Print out testing and training error
 
 Return the QDA object
 '''
def problem3():
    '''QDA test with testing data'''
    qda = QDA()
    qda_test_re = qda.fit(training_data,training_data_labels).score(testing_data,testing_data_labels)
    print("QDA prediction is:", qda.predict(testing_data) )
    qda_train_re = qda.fit(training_data,training_data_labels).score(training_data,training_data_labels)
    print("QDA testing error rate is", (1-qda_test_re)*100,"%")
    print("QDA training error rate is", (1-qda_train_re)*100,"%\n\n")
    return qda
```
